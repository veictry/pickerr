```python
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences

num_words = 10000
maxlen = 20
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)
word_index = imdb.get_word_index()

x_train2 = pad_sequences(x_train, maxlen=maxlen)
x_test2 = pad_sequences(x_test, maxlen=maxlen)
print(x_train2)
```

```python
reversed_word_index = dict([(key, word) for (word, key) in word_index.items()])
print(' '.join([reversed_word_index.get(key-3,'') for key in x_train[0]]))
print(y_train)
```

```python
from keras.layers import Flatten, Dense, Embedding
from keras.models import Sequential

model = Sequential()
model.add(Embedding(num_words, 8, input_length=maxlen))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy',
             metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train2, y_train, 
                   epochs=10,
                   batch_size=32,
                   validation_split=0.2)
```

```python
model.evaluate(x_test2, y_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.epoch, history.history['loss'])
```
